```r
# Load the data and display it
df <- read.csv("data_wt.csv")
df
```

```r
# Transform dimensions according to pixel size
px_size <- 0.06
df$x <- (df$x - min(df$x))*px_size
df$y <- (df$y - min(df$y))*px_size
```

```r
# Load library
library(MASS)

# Plot the time series belonging to both x and y axis
k <- kde2d(df$x, df$y)
image(k)
par(mfrow = c(2,1))
plot.ts(df$y)
plot.ts(df$x)
y <- df$x
if (max(df$y) > max(df$x))
{
        y <- df$y
}
```

```r
# Calculate the global spectrum and plot
par(mfrow = c(2,1))
del<-60 # sampling interval in seconds
y.spec <- spectrum(y,log="yes",span=2,plot=TRUE)

# Axis readjustment
spx <- y.spec$freq/del
spy <- 2*y.spec$spec
plot(spy~spx,xlab="frequency",ylab="spectral density",type="l")

# Ignore frequencies with a period greater than 16 minutes (too big)
cat(1/(peaks[peaks > 0.001]*60), sep = "\n")
```

```r
# Calculate the amplitude plots for the trajectories
interval = 4
y_d <- diff(y)
amplitude_plot <- as.numeric()
for (i in seq(from=interval, to=length(y_d), by=interval)) {
    amplitude_plot[i] <- sum(abs(y_d[(i-interval+1):i]))
}
amplitude_plot <- amplitude_plot[!is.na(amplitude_plot)]
par(mfrow = c(1,2))
plot(amplitude_plot)
plot(density(amplitude_plot)$y, density(amplitude_plot)$x, type = "l")
```

```r
# Distributions for the morphologies
par(mfrow = c(2,2))
plot(density(df$ecc))
plot(density(df$ep))
plot(density(df$mass))
plot(density(df$raw_mass))
```

```r
# Segmentation according to the windowed amplitude
library(segclust2d)
tstodf <- function(timeserie) {
  return(data.frame(Y=as.matrix(timeserie)))
}
segments <- (segmentation(df, lmin = 6, Kmax = 10, seg.var = c("y"), subsample_by = 1, scale.variable = TRUE))
plot(segments)
best_case <- segments$outputs$`4 segments`$segments
best_case$end - best_case$begin

# As observed, the segment 1 (red) contains the region which, in
# the upper spectrogram, contains significant frequencies, and those
# with a lesser significant slow-down towards the end. This is
# considered as the medium stage (displacement reduced, but period
# is observed)
```

```r
# Load the spectrum library
library(dplR)

# Load colors for the spectrogram
specCols <- c("#5E4FA2", "#3288BD", "#66C2A5", "#ABDDA4", "#E6F598", 
              "#FEE08B", "#FDAE61", "#F46D43", "#D53E4F", "#9E0142")

# Create the spectrogram
y <- df$y
wave.out <- morlet(y, 1:length(y), p2 = 8, dj = 0.1, siglvl = 0.999) 
wavelet.plot(wave.out, reverse.y = TRUE, key.cols=specCols)

# Lines are automatic for the previous segments, 73 is the end of
# the high stage
abline(v=c(best_case$end, best_case$begin), lwd = 1.5, col= c("blue","red"), lty=2)

# Find frequency minima above 7.5 minutes
tt <- rowSums(wave.out$Power[,1:30])
minima <- which(diff(sign(diff(tt)))==2)
abline(v=minima, lty = 2, col = "green", lwd = 1.5)
```

```r
# Load motif finding libraries
library(tsmp)

# Matrix algorithm for the Y axis
matrix <- tsmp(df$y, window_size = 12, s_size = 50, mode = "stomp")
motifs <- find_motif(matrix, n_motifs = 6)
plot(motifs)

# For global motif finding, please attach all time series 
# (this is an example)
```

```r
# Example of the observed appearance of a "rare" motif (stopping)
# versus a minimum value of convexity
par(mfrow = c(2,1))
plot.ts(df$y)
conv_min <- df$time[diff(df$ecc) <= quantile(diff(df$ecc), 0.01)]
abline(v=c(conv_min, conv_min + 20), col= "red", lty=2)
plot.ts(diff(df$ecc))
abline(h=quantile(df$ecc, 0.01), v=conv_min, col= "red", lty=2)

# Library loading
library(lmtest)

# Causation calculation
grangertest(diff(df$y), diff(df$ecc))$Pr
# As observed, minimum 0.01 quantile values in convexity
# (Exemplified) cause trajectory alterations
```
